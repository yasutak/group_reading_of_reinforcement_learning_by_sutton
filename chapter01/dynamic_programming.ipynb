```julia
# naive recursive algorithm for calculating fibonacci number

function fib(n)
    if n <= 2
        return 1
    else
        f = fib(n-1) + fib(n-2)
    end
    return f
end
```

```julia
for i in 1:10
    println(fib(i))
end
```

```julia
using BenchmarkTools, Plots
```

```julia
execution_times = []
for i in 1:10
    append!(execution_times, @belapsed fib($i))
end
```

```julia
#this implementation takes exponential time!
plot(execution_times)
```

```julia
#memorized DP algorithm for calculating fibonacci number

global memo = Dict()

function fib2(n)
    
    if haskey(memo, n)
        return memo[n]
    elseif n <= 2
        f = 1
    else
        f = fib2(n-1) + fib2(n-2)
    end
    
    memo[n] = f
    return f
end
```

```julia
execution_times2 = []
for i in 1:10
    append!(execution_times2, @belapsed fib2($i))
end
```

```julia
execution_times2
```

```julia
plot(execution_times)
plot!(execution_times2)
```

fib2 only takes linear time!
